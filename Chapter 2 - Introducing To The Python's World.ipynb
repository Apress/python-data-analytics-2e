### Functional Programming

```python
items = [1, 2, 3, 4, 5]
def inc(x): return x+1
```

```python
list(map(inc, items))
```

```python
list(map((lambda x: x + 1), items))
```

```python
list(filter((lambda x: x < 4), items))
```

```python
from functools import reduce
reduce((lambda x,y: x/y), items)
```

```python
S = [x**2 for x in range(5)]
S
```
